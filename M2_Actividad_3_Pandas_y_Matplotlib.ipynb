```python
import pandas as pd
import matplotlib as mpl
from google.colab import drive
drive.mount('/content/gdrive')
df = pd.read_csv('/content/gdrive/MyDrive/DATASET/ALPHABET')
```

```python
date1 = '24/08/20'
date2 = '02/09/20'

mpl.pyplot.plot(df.iloc[df.index[df['Date'] == date1][0]:df.index[df['Date'] == date2][0]+1,0],df.iloc[df.index[df['Date'] == date1][0]:df.index[df['Date'] == date2][0]+1,4])
mpl.pyplot.title("Date vs. Close")
mpl.pyplot.xlabel("Date (DD/MM/YY)")
mpl.pyplot.ylabel("Close (USD)")
```
